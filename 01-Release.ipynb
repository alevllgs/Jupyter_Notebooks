```python
lista_enteros = [entero for entero in range(100)] # lista con los 100 primeros números enteros
print(lista_enteros)
lista_pares = []
lista_impares = list()
```

### Objetivo
Separar esa lista_enteros en 2 sublistas, una con los números pares y otra con los impares.

Recordar que el operador _`%`_ retorna el resto de la división entera.

También recordar que los números pares se pueden definir como 
los números cuyo resto de la division entera por 2 es 0.

```python
for i in lista_enteros:
    if i%2 == 0:
        lista_pares.append(i)
    else:
        lista_impares.append(i)
print(lista_pares)
print(lista_impares)
```

### Mini Ejercicio 2

```python
lista_nombres = ['Juan', 'Pedro', 'Thomas', 'Juana', 'Maria']
lista_ruts = ['92134518-k', '62754118-k', '55144322-k', '91456512-k', '999888777-k']
```

Tenemos dos listas, una con los nombres de cinco personas y otra con la rut de esas personas, queremos unir la información de esas listas de tal forma que el resultado final sea una lista de tuplas, donde cada elemento es una tupla de la forma `('Nombre_Persona_1, 'rut_persona_1')`.
Ambas listas estan ordenadas, así que el primero elemento de la lista de nombre corresponde a la rut del primero elemento de la lista de rut´s y eso se cumple para todos los elementos.

```python
lista_de_tuplas=list()
for i in range(len(lista_nombres)):
    lista_de_tuplas.append((lista_nombres[i],lista_ruts[i]))
print(lista_de_tuplas)
```
