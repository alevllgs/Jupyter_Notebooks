# Clases


*   **Clase o instancia** permite que la creación y el manejo de los objetos se 
sencillo, es una plantilla que nos indica cuales deben ser las caracteristicas y acciones que debe hacer cada objeto.

*   **Objeto** puede ser un ojeto real como una cuenta, un apellido, un número de serie o abstracto como una ecuación, una fracción o un número complejo. Cada objeto tiene un conjunto de características que lo definen asi como un conjunto de acciones que determinan su comportamiento con respecto a otros objetos.


*   **Atributo** es el valor que almacena internamente el objeto (las caracteristicas del objeto) y que pueden ser datos u otros objetos.
*   **Método** es un grupo de instrucciones asociadas al objeto a las que se les ha dado un nombre en especifico. Cuando un metodo es invocado se ejecutaan las instrucciones del programa. Los métodos pueden acceder y modificar a los atributos que contiene el objeto (una función no lo puede hacer)


*   **Encapsular** un objeto quiere decir que se puede almacenar informacion y trabajar con ella de tal manera que los cambios de estado solamente pudan ser sealizados por los métodos que prtenecen al objeto y que ningun otro metodo o función externa puedan hacerlo.

*   **Herencia** es una técnica de la POO en la cual se puede crear una clase nueva a través de una clase existente, por medio de la declaracion de nuevos métodos que  y atributos que extiendan la funcionalidad de la clase de la que se hereda. Los objetos de las nuevas clases heredan las propiedades y comportamientos de todas las clases a las que pertenecen.


*   **Polimorfismo** es la característica de la POO que permite a los valores de diferentes tipos de datos, ser manejados usando una interfaz uniforme, el mismo método puede ser usado por varias clases, pero de forma diferente.



























**class Estudiantes:**  en esta parte definimos una variable

**def __init__(self, n, a, e):**  este método se llama **inicializador** y se ejecuta cada vez que creamos un objeto perteneciente a a esta clase


```python
class Estudiantes:

  def __init__(self, n, a, e):

    self.nombre = n
    self.apellido = a
    self.edad = e
    self.lista_de_notas = [] #este parametro no lo definimos en el init, por lo que es igual para todos

  def __str__(self):
      return f"Estudiantes {self.nombre} {self.apellido} de {self.edad} años. Su nota es: {self.lista_de_notas} y su promedio es {self.promedio()}"

  def agregar_nota(self, nueva_nota):
    self.lista_de_notas.append(nueva_nota)

  def subir_puntajes(self, x):
    for i in range(len(self.lista_de_notas)):
      self.lista_de_notas[i] += x
      if self.lista_de_notas[i] > 7.0:
        self.lista_de_notas[i] = 7.0

  def promedio(self):
    if len(self.lista_de_notas) == 0:
      return 0.0
    suma = 0  #variable local, solo existe dentro del método, no se guarda
    for nota in self.lista_de_notas:
      suma += nota
    return suma / len(self.lista_de_notas)
```

Tanto g como h son **objetos** pertenecientes a la clase **Estudiantes:** ⏬

```python
g = Estudiantes("Ana", "Pérez", 35)
h = Estudiantes("Juan", "Soto", 36)

## Accion 1. Agregarle notas a cada estudiante. (se vincula con def agregar_nota)
g.agregar_nota(3.3)
g.agregar_nota(6.3)
g.agregar_nota(2.4)
h.agregar_nota(5.1)
h.agregar_nota(6.1)

## Mostrar el estado de cada estudiante (se vincula con def __str__(self))
print(g)
print(h)
print()

# Acción 2. Subirle 1.5 puntos a cada nota. (se vincula con def subir_puntaje)
g.subir_puntajes(1.5)
h.subir_puntajes(1.5)

## Mostrar el estado de cada estudiante (se vincula con def __str__(self))
print(g)
print(h)
print()

# Acción 3. Calcular el promedio de cada estudiante. (se vincula con def promedio) (se vincula con def __str__(self))
print(g)
print(h)
print()
```

## Herencia

Le **herencia** permite reutilizar y extender clases previamente creadas, lo cual permite la reutilización del código, siendo no necesario escribir métodos o atributos que son creados en las clases de las cuales se hereda.

```python
class Persona:
  def __init__(self,nombre, apellido, edad):
    self.nombre = nombre
    self.apellido = apellido
    self.edad = edad

    def subir_edad(self, anios):
      self.edad += anios

class Estudiante(Persona):
  def __init__(self, nombre, apellido, edad):
    super().__init__(nombre, apellido, edad)
    self.notas = []

    def agregar_notas(self, nota):
      self.notas.append(nota)
```

```python
g = Estudiante("Marco", "Fredes", 37)
print(g.nombre, g.apellido, g.notas)
```
