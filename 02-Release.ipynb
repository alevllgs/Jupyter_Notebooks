Debes crear la clase Persona, la cual contiene los atributos nombre y apellido

```python
#Ya esta listo, avanzar para la parte 2
class Persona:
    def __init__(self, nombre, apellido):
        self.nombre = nombre
        self.apellido = apellido
    
    def __repr__(self):
        return f'hola soy {self.nombre} {self.apellido}'
```
